Part A Count the top ten unique words in Little Women

```python
#count the the top ten unique words without filters
!cat women.txt | grep -oE '\w{{1,}}' | tr '[:upper:]' '[:lower:]' | sort | uniq -c | sort -rn | head -10
```

```python
# count the top ten unique words with filters
# Split.py splits lines of text into one word per line
# lowercase.py transforms text into lower case
```

```python
!chmod +x Split.py
!chmod +x lowercase.py
```

```python
!cat women.txt | ./Split.py | ./lowercase.py | sort | uniq -c | sort -rn | head -10
```

Part B (10 points)
Write a Python filter that removes at least ten common words of English text, commonly known as “stop words”. 
Add your stop word filter to a word count pipeline and show the top 25 words in Little Women with stop words removed.

```python
#stop.py removes fifteen common words of English text. Resource:https://en.wikipedia.org/wiki/Most_common_words_in_English
!chmod +x stop.py
```

```python
!cat women.txt|  grep -oE '\w{{1,}}' |  ./lowercase.py |  ./stop.py | sort | uniq -c | sort -rn | head -25
```
