How many insepections are given to each owner?

```python
%%sql
SELECT OWNER_NAME, COUNT(OWNER_NAME) AS number_of_inspections 
FROM food
GROUP BY OWNER_NAME
ORDER BY COUNT(OWNER_NAME) DESC;
```

I find APS gets most inspections and let's take a close look.

```python
%%sql
SELECT INSPECTION_DESC,
COUNT(INSPECTION_DESC) FROM food AS food
WHERE OWNER_NAME = 'APS'
GROUP BY "INSPECTION_DESC"
ORDER BY COUNT(INSPECTION_DESC) DESC;
```

From above, it can be found that APS gets a large number of routine food inspections and it also gets many new business inspection and pre-opening food inspections. It appears that APS is newly opened and that may be the reason why it gets most inspections.

What are different types of violation and which one is the most popular one?

```python
%%sql
SELECT DISTINCT VIOLATION_CODE, COUNT(*)
FROM food
GROUP BY VIOLATION_CODE
ORDER BY COUNT(*) DESC;
```

It can be found that 55 is the most frequent violation type.

How many different types of inspections are taken each day per type?

```python
%%sql
SELECT INSPECTION_DATE, INSPECTION_DESC, COUNT(INSPECTION_DESC)
FROM food
GROUP BY INSPECTION_DATE;
```

I would like to normalize violation code and violation description.

First, I create a table named violations.

```python
%%sql
DROP TABLE IF EXISTS violations;
CREATE TABLE violations
(code int primary key,
description varchar(64)
);
```

```python
%%sql
INSERT INTO violations (code)
SELECT DISTINCT CAST("VIOLATION_CODE" as int) FROM food;
```

```python
%%sql
UPDATE violations as v
SET description = f."VIOLATION_DESC"
FROM food as f
WHERE code = CAST(f.VIOLATION_CODE AS int);
```
